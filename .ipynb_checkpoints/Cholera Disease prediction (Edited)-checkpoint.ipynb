```python
cholera_data.drop(columns = 'WHO Region', axis=1, inplace=True)
```

```python
#this separates the target column from other columns
x = cholera_data.drop(columns='Cholera case fatality rate', axis=1)
y = cholera_data['Cholera case fatality rate']
print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

x.Country = le.fit_transform(x.Country)
```

```python
#splitting the data into training and test data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

**Import Regression Algorithms**

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
KNC = KNeighborsRegressor()
SVR = SVR()
LRG = LinearRegression()
```

```python
x_train
```

```python
#here we fit our training dataset into the model
KNC.fit(x_train, y_train)
SVR.fit(x_train, y_train)
LRG.fit(x_train, y_train)
```

```python
#prediction for the test data (x_test)
KNC_prediction = KNC.predict(x_test)
LRG_prediction = LRG.predict(x_test)
SVR_prediction = SVR.predict(x_test)

#prediction for the training data (x_train)
KNC_x_train_prediction = KNC.predict(x_train)
LRG_x_train_prediction = LRG.predict(x_train)
SVR_x_train_prediction = SVR.predict(x_train)
```

**Classification Evaluation Metrics**

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
mse_KNC = metrics.mean_squared_error(y_test, KNC_prediction)
mse_LRG = metrics.mean_squared_error(y_test, LRG_prediction)
mse_SVR = metrics.mean_squared_error(y_test, SVR_prediction)
```

```python
#print Error score for training dataset
print('MSE on training data for KNearestNeigbour: ', mean_squared_error(y_train, KNC_x_train_prediction))
print('MSE on training data for LinearRegression: ', mean_squared_error(y_train, LRG_x_train_prediction))
print('MSE on training data for SVR: ', mean_squared_error(y_train, SVR_x_train_prediction))
```

```python
#print MSE score for test dataset
print('Test MSE for KNearestNeigbour: ', mse_KNC)
print('Test MSE for LogisticRegression: ', mse_LRG)
print('Test MSE for SVR: ', mse_SVR)
```

```python
rmse_KNC = np.sqrt(mse_KNC)
rmse_LRG = np.sqrt(mse_LRG)
rmse_SVR = np.sqrt(mse_SVR)
```

```python
#print RMSE score for test dataset
print('Test RMSE for KNearestNeigbour: ', rmse_KNC)
print('Test RMSE for LogisticRegression: ', rmse_LRG)
print('Test RMSE for KNearestNeigbour: ', rmse_SVR)
```

## **Ensemble Method**


**Import Classifier Boosting Algorithms**

```python
from sklearn.ensemble import GradientBoostingRegressor
from catboost import CatBoostRegressor
```

```python
#We now create an object of class LogisticRegression() to train the model on
CBC = CatBoostRegressor()
GBC = GradientBoostingRegressor()
```

```python
#here we fit our training dataset into the model
CBC.fit(x_train, y_train)
GBC.fit(x_train, y_train)
```

```python
#prediction for the test data (x_test)
CBC_prediction = CBC.predict(x_test)
GBC_prediction = GBC.predict(x_test)

#prediction for the training data (x_train)
CBC_x_train_prediction = CBC.predict(x_train)
GBC_x_train_prediction = GBC.predict(x_train)
```

```python
mse_GBC = metrics.mean_squared_error(y_test, GBC_prediction)
mse_CBC = metrics.mean_squared_error(y_test, CBC_prediction)
```

```python
#print Error score for training dataset
print('MSE on training data for GradientBoost: ', mean_squared_error(y_train, GBC_x_train_prediction))
print('MSE on training data for CatBoost: ', mean_squared_error(y_train, CBC_x_train_prediction))
```

```python
#print MSE score for test dataset
print('Test MSE for GradientBoost: ', mse_GBC)
print('Test MSE for CatBoost: ', mse_CBC)
```

```python
rmse_GBC = np.sqrt(mse_GBC)
rmse_CBC = np.sqrt(mse_CBC)
```

```python
#print RMSE score for test dataset
print('Test RMSE for GradientBoost: ', rmse_GBC)
print('Test RMSE for CatBoost: ', rmse_CBC)
```

**Stacking**

```python
# Define estimators
from sklearn.ensemble import StackingRegressor
estimators = [
    ('KNC', KNC),
    ('GBC', GBC),
    ('CBC', CBC)
]
# Build stack model
stack_model = StackingRegressor(estimators = estimators, final_estimator = LinearRegression())
# train stacked model
stack_model.fit(x_train,y_train)
#make predictions
y_train_prediction = stack_model.predict(x_train)
y_test_prediction = stack_model.predict(x_test)
#Train data model performance
stack_model_train_mse= metrics.mean_squared_error(y_train, y_train_prediction)
#Test data model performance
stack_model_test_mse= metrics.mean_squared_error(y_test, y_test_prediction)
```

```python
print('\n',"Stack model performance on the Training set")
print('stack_model_train_MSE: ', stack_model_train_mse)
print("Stack model performance on the Testing set")
print('stack_model_test_MSE: ', stack_model_test_mse)
print('\n'*2)
print('stack_model_test_RMSE: ', np.sqrt(stack_model_test_mse))
```

## Let's Visualize and Compare the Performance of all the Models.

```python
acc = []
model = []

#Let us store the models and their accuracy score in the empyt lists we created

acc.append(rmse_CBC)
model.append('Catboost ')

acc.append(rmse_KNC)
model.append('KNN ')

acc.append(rmse_SVR)
model.append('SVM ')

acc.append(rmse_LRG)
model.append('Linear Regression ')

acc.append(rmse_GBC)
model.append('Gradient Boost ')

acc.append(np.sqrt(stack_model_test_mse))
model.append('Stacking Classifier ')

plt.figure(figsize=[10,5],dpi = 100)
plt.title('Accuracy Comparison for Result')
plt.ylabel('Accuracy')
plt.xlabel('Algorithm')
sns.boxplot(y = acc,x = model,palette='dark')
plt.show()
```

```python
"""
Created on Sat Aug 26 05:24:33 2023

@author: YASHIM GABRIEL
"""

import pickle
import gradio as gr


model = pickle.load(open('Catboost_classifier.pkl', 'rb'))


def crop(Country, Year, Cases, Deaths):
    prediction = model.predict([[Country, Year,Cases, Deaths]])
    return prediction[0]


# create input and output objects
# input object1
input1 = gr.inputs.Dropdown(
            ["ran", "swam", "ate", "slept"], value=["swam", "slept"], multiselect=False, label="Country" info="Dropdown to select Country"
        )
# input object 2
input2 = gr.inputs.Number(label="Year)")
# input object3
input3 = gr.inputs.Number(label="Number of reported cases of cholera")
# input object 4
input4 = gr.inputs.Number(label="Number of reported deaths from cholera")

# output object
output = gr.outputs.Textbox(label="Cholera case fatality rate")


# create interface
gui = gr.Interface(fn=crop,
                   inputs=[input1, input2, input3,
                           input4],
                   outputs=output, title='Cholera case fatality rate Prediction System', description=' ').launch()
```
